1. Measure accuracy of your model using different kernels such as rbf and linear.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Importing the data set from matplot library

from sklearn.datasets import load_digits
digits = load_digits()
digits.target
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
df = pd.DataFrame(digits.data,digits.target)
df.head()
```

```python
df['target'] = digits.target
df.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)
```

```python
len(x_train), len(x_test)
```

## Kernel Optimization

Types of kernals: 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'

```python
## Using Linear kernel

from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
linear_model = SVC(kernel='linear')
linear_model.fit(x_train,y_train)
linear_model.score(x_test,y_test)
```

```python
## Using poly kernel

from sklearn.svm import SVC
rbf_model = SVC(kernel='poly')
rbf_model.fit(x_train, y_train)
rbf_model.score(x_test, y_test)
```

```python
## Using RBF kernel

from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
rbf_model.fit(x_train, y_train)
rbf_model.score(x_test, y_test)
```

Note: RBF and Poly kernels may have higher accuracy, but the linear kernel might be much more suitable in this case as the accuracy are very near.
