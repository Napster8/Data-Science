## Importing and Anatomy

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Method 1: Less flexible (not recommended). 

x = [1, 2, 3, 4]
y = [10, 20, 35, 100]
plt.plot(x, y);
```

```python
# Method 2: Sub-plots leave us room to plot more graphs in the same inline (recommended)

fig, ax = plt.subplots()
ax.plot(x,y);
```

## Line Graph

```python
# 1. Import matplotlib
# already done at the top of the book

# 2. Prepare or import data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 3. Setup plot
fig, ax = plt.subplots(figsize = (3, 3)) # (width, height) of the plot

# 4. Customize plot
ax.plot(x, y)
ax.set(title = "Simple Plot",
      xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save as image
fig.savefig("../images/sample_plot.png")
```

## Bar Plot

```python
# Creating dictionary of prices. Plot to depict what cost comparison.
# "Key1": Value1, "Key2": Value2...

prices = {"Almond": 10, "Peanut": 8, "Cashew": 12}  

fig, ax = plt.subplots()
ax.bar(prices.keys(), prices.values())
ax.set(title="Price Comparison", ylabel="Price ($)");
```

## Horizontal Bar Plot

```python
fig, ax = plt.subplots()

# turn the keys and values into a list
ax.barh(list(prices.keys()), list(prices.values()))
ax.set(title="Price Comparison", xlabel="Price ($)");
```

## Scatter Plot

```python
#data

x = np.random.randint(0, 50, 50)
y = np.random.randint(0, 50, 50)
 
#scatter plot

fig, ax = plt.subplots()
ax.scatter(x,y)
ax.set(title = "Scatter Plot", xlabel = "x-axis", ylabel = "y-axis");
```
