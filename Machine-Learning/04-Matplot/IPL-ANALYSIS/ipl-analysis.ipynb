```python
# Libraries and dependencies for basic visuals and analysis.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ipl = pd.read_csv('ipl_data.csv')
ipl.columns = ipl.columns.str.replace(' ', '')
ipl.head()
```

* It seems that there is no need to clean the  dataset.

```python
ipl1 = ipl.groupby(['Team']).sum()
ipl1
```

```python
ipl2 = ipl1.drop(ipl1.columns[[0, 1, 7, 8, 9 , 10, 11, 12, 13, 13]], axis = 1)
ipl2
```

```python
ipl2.index.name = 'Team'
ipl2.reset_index(inplace=True)
ipl2
```

## Won/Loss (wl) Ratio.

* True measure of performance may not be the number of matches, the team has played.
* The number of matches won may not be the true performance measure either as the teams have played different number of matches.
* So we need to create a Won/loss Ratio. Higher this ratio, the better would be the performance of a team.

```python
ipl2['wl'] = ipl2['Won'] / ipl2['Loss']
ipl2
```

### WIN/LOSS RATIO

```python
# sort and plot
ipl2 = ipl2.sort_values(by = 'wl', ascending = False, inplace = False)

plt.figure(figsize = (10, 5))
sns.barplot(x = 'wl', y = 'Team', data = ipl2, orient = 'h')
plt.xlabel("Win/Loss Ratio")
plt.ylabel("Teams Participated");
```

### TOTAL POINTS

```python
# sort and plot

ipl2 = ipl2.sort_values(by = 'Points', ascending = False, inplace = False)

plt.figure(figsize = (10, 5))
sns.barplot(x = 'Points', y = 'Team', data = ipl2, orient = 'h')
plt.xlabel("Total Points")
plt.ylabel("Teams Participated");
```

```python
# add wl column
ipl['wl'] = ipl['Won'] / ipl['Loss']
```

## Yearly Trend of specified team

```python
def yearly_trend(team_name):
    plt.figure(figsize = (10, 5))
    filtered_results = ipl[(ipl['Team']) == team_name]
    sns.barplot(x = filtered_results.Year, y = filtered_results.wl, data = filtered_results, orient='v')
    plt.xlabel(f"Evolution of Years for {team_name}")
    plt.ylabel(f"Win/Loss Ratio of {team_name}")
yearly_trend('KingsXIPunjab') 
```

```python
yearly_trend('RajasthanRoyals')
```

```python
yearly_trend('ChennaiSuperKings')
```

## Performance of all the teams in a specified year

```python
def performance_in(year):
    plt.figure(figsize = (10, 5))
    filtered_results = ipl[(ipl['Year']) == year]
    sns.barplot(y = filtered_results.Team, x = filtered_results.wl, data = filtered_results, orient='h')
    plt.ylabel(f"Teams performance in {year}")
    plt.xlabel(f"Win/Loss Ratio of {year}")

performance_in(2017)
```
